```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

data = pd.read_csv('train.csv')

print(data.head())
print(data.isnull().sum())
data.columns
```

```python
# выжил или нет
plt.figure(figsize=(6, 5))
plt.hist(data['Survived'], bins=[0, 1, 2], color='deepskyblue', rwidth=0.8)
plt.title('Survived')
plt.ylabel('Count')
plt.xticks([0.5, 1.5], ['died', 'survived'])
plt.show()
```

```python
# Какой пол был у погибших/выживших

fig, ax = plt.subplots(figsize=(6, 5))


men = data[data['Sex'] == 'male']['Survived']
women = data[data['Sex'] == 'female']['Survived']
ax.hist([men, women], bins=[0, 1, 2], stacked=False, color=['skyblue', 'deeppink'])

ax.legend(['Male', 'Female'])

ax.set_title('Survived by sex')
ax.set_xlabel('Survived')
ax.set_ylabel('Count')
ax.set_xticks([0.5, 1.5])
ax.set_xticklabels(['died', 'survived'])

plt.show()
```

```python
# Какой класс был у погибших/выживших

fig, ax = plt.subplots(figsize=(6, 5))


first = data[data['Pclass']==1]['Survived']
second = data[data['Pclass']==2]['Survived']
third = data[data['Pclass']==3]['Survived']
ax.hist([first, second, third], bins=[0, 1, 2],  stacked=False, color=['thistle', 'plum', 'purple'])

ax.legend(['1 class', '2 class', '3 class'])

ax.set_title('Survived by class')
ax.set_xlabel('Survived')
ax.set_ylabel('Count')
ax.set_xticks([0.5, 1.5])
ax.set_xticklabels(['died', 'survived'])

plt.show()
```

```python
#  Какого возраста были пошибшие/выжившие

fig, ax = plt.subplots(figsize=(6, 5))


ten = data[(data['Age']>=0) & (data['Age']<=10)]['Survived']
twenty = data[(data['Age']>=11) & (data['Age']<=20)]['Survived']
thirty = data[(data['Age']>=21) & (data['Age']<=30)]['Survived']
forty = data[(data['Age']>=31) & (data['Age']<=40)]['Survived']
fifty = data[(data['Age']>=41) & (data['Age']<=50)]['Survived']
sixty = data[(data['Age']>=51) & (data['Age']<=60)]['Survived']
seventy = data[(data['Age']>=61) & (data['Age']<=70)]['Survived']
eighty = data[(data['Age']>=71) & (data['Age']<=80)]['Survived']


ax.hist([ten, twenty, thirty, forty, fifty, sixty, seventy, eighty], bins=[0, 1, 2],
  stacked=False, color=['thistle', 'plum', 'purple', 'darkmagenta', 'darkslateblue', 'rebeccapurple', 'blueviolet', 'indigo'])

ax.legend(['0-10', '11-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80'])

ax.set_title('Survived by age')
ax.set_xlabel('Survived')
ax.set_ylabel('Count')
ax.set_xticks([0.5, 1.5])
ax.set_xticklabels(['died', 'survived'])

plt.show()
```

```python
#  Сколько было братьев/сестер у погибших/выживших

fig, ax = plt.subplots(figsize=(6, 5))


zero = data[data['SibSp']==0]['Survived']
one = data[data['SibSp']==1]['Survived']
two = data[data['SibSp']==2]['Survived']
three = data[data['SibSp']==3]['Survived']


ax.hist([zero, one, two, three], bins=[0, 1, 2],  stacked=False, color=['thistle', 'plum', 'purple', 'rebeccapurple'])

ax.legend(['0 siblings', '1 sibling', '2 siblings', '3 siblings'])

ax.set_title('Survived with siblings')
ax.set_xlabel('Survived')
ax.set_ylabel('Count')
ax.set_xticks([0.5, 1.5])
ax.set_xticklabels(['died', 'survived'])

plt.show()
```

```python
# сколько заплатили выжившие/погибшие за билеты
fig, ax = plt.subplots(figsize=(6, 5))


cheap = data[data['Fare']<=25]['Survived']
middle = data[(data['Fare']>25) & (data['Fare']<=50)]['Survived']
high = data[(data['Fare']>50) & (data['Fare']<=100)]['Survived']
expensive = data[data['Fare']>100]['Survived']

ax.hist([cheap, middle, high, expensive], bins=[0, 1, 2],  stacked=False, color=['thistle', 'plum', 'purple', 'rebeccapurple'])

ax.legend(['< 25', '25-50', '50-100', '> 100'])

ax.set_title('Survived by ticket price')
ax.set_xlabel('Survived')
ax.set_ylabel('Count')
ax.set_xticks([0.5, 1.5])
ax.set_xticklabels(['died', 'survived'])

plt.show()
```

```python
# В каком порту сели погибшие/выжившие
fig, ax = plt.subplots(figsize=(6, 5))


S = data[data['Embarked']=='S']['Survived']
C = data[data['Embarked']=='C']['Survived']
Q = data[data['Embarked']=='Q']['Survived']


ax.hist([S, C, Q], bins=[0, 1, 2],  stacked=False, color=['thistle', 'plum', 'purple'])

ax.legend(['S', 'C', 'Q'])

ax.set_title('Survived by seaport')
ax.set_xlabel('Survived')
ax.set_ylabel('Count')
ax.set_xticks([0.5, 1.5])
ax.set_xticklabels(['died', 'survived'])

plt.show()
```
